```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
```

```python
url="https://www.trulia.com/NY/New_York/"
```

```python
response=requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
result=soup.find_all(class_="Grid__CellBox-sc-144isrp-0")
```

```python
result
```

```python
len(result)
```

```python
# the pages shows only 30 results but we are getting 42. Some posts are hidden from the page

# filtering out those pages
results_update=[]

for i in result:
    if i.has_attr("data-testid"):
        results_update.append(i)
```

```python
len(results_update)
```

```python
#street
#region
#bedrooms
#bathrooms
#price
```

```python
results_update[0].find('div',{'data-testid':"property-street"})
```

```python
street=[]
region=[]
bedrooms=[]
bathrooms=[]
price=[]
for i in results_update:
    street.append(i.find('div',{'data-testid':"property-street"}).get_text())
    region.append(i.find('div',{'data-testid':"property-region"}).get_text())
    bedrooms.append(i.find('div',{'data-testid':"property-beds"}).get_text())
    bathrooms.append(i.find('div',{'data-testid':"property-baths"}).get_text())
    price.append(i.find('div',{'data-testid':"property-price"}).get_text())

   
```

```python
df=pd.DataFrame({"Street":street, 'Region':region,"Bedrooms":bedrooms,'Bathrooms':bathrooms,'Price':price})
```
