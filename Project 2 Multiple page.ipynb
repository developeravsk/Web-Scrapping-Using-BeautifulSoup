```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json




df=pd.DataFrame(columns=['Street','Region','Bedrooms','Bathrooms',"Price"])
```

```python
for i in range(1,3):
    url="https://www.trulia.com/NY/New_York/"+str(i)+"_p/"

    response=requests.get(url)
    soup=BeautifulSoup(response.content,'html.parser')
    result=soup.find_all(class_="Grid__CellBox-sc-144isrp-0")

    results_update=[]

    for i in result:
        if i.has_attr("data-testid"):
            results_update.append(i)

    street=[]
    region=[]
    bedrooms=[]
    bathrooms=[]
    price=[]
    for i in results_update:
        street.append(i.find('div',{'data-testid':"property-street"}).get_text())
        region.append(i.find('div',{'data-testid':"property-region"}).get_text())
        bedrooms.append(i.find('div',{'data-testid':"property-beds"}).get_text())
        bathrooms.append(i.find('div',{'data-testid':"property-baths"}).get_text())
        price.append(i.find('div',{'data-testid':"property-price"}).get_text())
    
    df = df.append(pd.DataFrame({'Street': street,'Region':region,'Bedrooms':bedrooms,'Bathrooms':bathrooms,'Price':price}), ignore_index=True)
```
