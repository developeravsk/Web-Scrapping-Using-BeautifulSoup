```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import urllib.parse

```

```python
url_2=[]
names=[]
address=[]
phones=[]

for num in range(1,4):
    print("page:",num)
    
    url="https://www.yellowpages.com/search?search_terms=restaurant&geo_location_terms=New%20York%2C%20NY&page="+str(num)
    response=requests.get(url)
    soup=BeautifulSoup(response.content,'html.parser')
    results=soup.find_all(class_="result")
    url_1="https://www.yellowpages.com/"
    for i in range(0,len(results)):
        print("data:",i)
        url_2.append(urllib.parse.urljoin(url_1,results[i].find(class_='business-name').get('href')))
        names.append(results[i].find(class_='business-name').get_text())
        address.append(results[i].find(class_='street-address').get_text())
        phones.append(results[i].find(class_='phones phone primary').get_text())
```

```python
website=[]
email=[]
for i in url_2:
    print("data:",i)
    response=requests.get(i)
    soup=BeautifulSoup(response.content,'html.parser')
    results=soup.find_all(class_="business-card-footer")
    try:
        website.append(results[0].find('a',{'class':"primary-btn website-link"}).get('href'))
    except:
        website.append('N/A')
    try:
        email.append(results[0].find('a',{'class':"email-business"}).get('href').strip('mailto:'))
    except:
        email.append('N/A')
```

```python
df=pd.DataFrame(columns=['Name','Address','Phone','Email',"HomePage"])
df = df.append(pd.DataFrame({'Name': names,'Address':address,'Phone':phones,'Email':email,'HomePage':website}), ignore_index=True)
```
